```python
import pandas as pd
import xml.etree.ElementTree as xmltree
import os
```

```python
hymns=pd.read_csv('hymns.csv')
hymns
```

```python
hymns=hymns.drop(["web-scraper-order","web-scraper-start-url","church-link-href","hymn-pagination","hymn-selector-href"],axis=1)
hymns
```

```python
def remove_unicode_characters(item):
    clean=item.encode("ascii","ignore").decode()
    return clean

hymns['church-link']=hymns['church-link'].apply(remove_unicode_characters)
hymns['hymn-selector']=hymns['hymn-selector'].apply(remove_unicode_characters)
hymns['hymn-title']=hymns['hymn-title'].apply(remove_unicode_characters)
hymns['hymn-lyrics']=hymns['hymn-lyrics'].apply(remove_unicode_characters)
```

```python
def GenerateXML(hymn) :
      
#  <song xmlns="http://openlyrics.info/namespace/2009/song"
#       version="0.9">
#   <properties>
#     <titles>
#       <title>Amazing Grace</title>
#     </titles>
#   </properties>
#   <lyrics>
#     <verse name="v1">
#       <lines>
#         Amazing grace how sweet the sound
#       </lines>
#     </verse>
#   </lyrics>
# </song>
    root = xmltree.Element("song")
    root.set("xmlns","http://openlyrics.info/namespace/2009/song")
    root.set("version","0.9") 

    properties = xmltree.Element("properties")
    root.append (properties)
      
    titles = xmltree.SubElement(properties, "titles")

    title = xmltree.SubElement(titles,"title")
    title.text = hymn['hymn-selector'].to_string()

    authors = xmltree.SubElement(properties,"authors")
    author = xmltree.SubElement(authors,"author")
    author.text=hymn['church-link'].to_string()

    lyrics = xmltree.SubElement(root,"lyrics")
    
    # Splitting hymns into verses
    verse_number=1
    lines_from_hymn=hymn['hymn-lyrics'].to_string().strip().split("\\n")
    print(lines_from_hymn)
    print(len(lines_from_hymn))

    for line in lines_from_hymn:
        if line != '':
            verse = xmltree.SubElement(lyrics,"verse")
            verse.set("name",verse_number)
            lines = xmltree.SubElement(verse,"lines")
            lines.text=line
            verse_number+=1

    tree = xmltree.ElementTree(root)
      
    with open (f"{hymn['hymn-title'].to_string()}.xml", "wb") as files :
        tree.write(files)

```
