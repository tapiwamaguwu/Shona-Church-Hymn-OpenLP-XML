```python
# removes unicode chararcters since the file has a lot of dirty characters
def clean_up_unicode(dirty:str):
    return dirty.encode("ascii","ignore").decode()
```

```python
# <lyrics> <verse name=\"v1\"> <lines> Amazing grace how sweet the sound </lines> </verse> </lyrics>
# generates xml formatted lyrics as the example above
def get_lyrics_xml(lyrics):
    lyrics=lyrics.encode("ascii","ignore").decode()
    verse_number=1
    lyrics_xml_output="<lyrics>"
    for lyric in lyrics.split("<EOV><EOV><EOV><EOV>"):
        if lyric.strip()!='':
            lyric=lyric.replace("<EOV><EOV>","\n")
            i=f"<verse name=\"v{verse_number}\"> <lines> {lyric} </lines> </verse>"
            lyrics_xml_output+=i
            verse_number+=1
    lyrics_xml_output+="</lyrics>"
    return lyrics_xml_output  
```

```python
#  write get whole song as xml (adds other song metadata like author, title, and uses the get_lyrics function to add the lyrics in xml format)
def get_song_xml(hymn_dict):
    xml_text=f"<?xml version=\"1.0\" encoding=\"UTF-8\"?><song xmlns=\"http://openlyrics.info/namespace/2009/song\" version=\"0.9\"> <properties> <titles> <title>{hymn_dict['hymn-title']}</title> <title>{clean_up_unicode(hymn_dict['hymn-selector'])}</title> </titles> <authors><author>{hymn_dict['church-link']}</author></authors> </properties> {get_lyrics_xml(hymn_dict['delimited-hymn-lyrics'])} </song>"
    return xml_text
```

```python
# Uses text to manipulation to create an xml file
def generate_xml(hymn):
    song_xml=get_song_xml(hymn)
    with open(f"output/{hymn['hymn-title']}.xml","w") as f:
        f.write(song_xml)
```

```python
# Iterate through the DataFrame and generate an XML file for each row
hymns.apply(generate_xml,axis=1)
```
